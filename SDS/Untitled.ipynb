```python
import requests
from bs4 import BeautifulSoup
```

```python
link="https://www.amazon.in/Apple-iPhone-XR-64GB-Black/product-reviews/B07JWV47JW"
```

```python
page=requests.get(link)
```

```python
print(page)
```

```python
soup=BeautifulSoup(page.content,'html.parser')
```

```python
print(soup)
```

```python
print(soup.prettify())
```

```python
name=soup.find_all(class_="a-profile-name")
print(name)
```

```python
name=soup.find_all(class_="a-profile-name")
name
```

```python
cust_name=[]
for i in range(0,len(name)) :
    cust_name.append(name[i].get_text())
    
cust_name
```

```python
review_title=soup.find_all(class_="a-size-base a-link-normal review-title a-color-base review-title-content a-text-bold")
review_title
```

```python
cust_review_title=[]
for i in range(0,len(review_title)) :
    cust_review_title.append(review_title[i].get_text())
    
cust_review_title
```
